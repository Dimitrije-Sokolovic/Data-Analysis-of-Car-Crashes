```python
tip = dfsrb['broj_vozila'].value_counts()

fig = px.bar(x=tip.index,
            y=tip,
            title='Tip saobracajne nesrece',
            height=1000,
            text=dfsrb['broj_vozila'].value_counts()/len(dfsrb['broj_vozila'])*100)

fig.update_traces(texttemplate='%{text:.4s}%',
                 textposition='outside',
                 marker=dict(color='magenta',line=dict(color='tomato')))

fig.update_layout(margin={'b':0,'l':0,'r':0,'t':100},
                 paper_bgcolor='rgb(248, 248, 255)',
                 plot_bgcolor='rgb(248, 248, 255)',
                 title={'font':{
                             'family':'monospace',
                             'size':22,
                             'color': 'grey'
                         },
                        'x':0.45,'y':0.9})
fig.update_xaxes(
        tickangle = 60,
        title_font = {"size": 20},
        title_standoff = 25)
fig.show()
```
